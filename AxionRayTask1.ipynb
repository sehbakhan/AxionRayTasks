```python
plt.figure(figsize=(12, 6))  
sns.countplot(y='Root Cause', data=df, order=df['Root Cause'].value_counts().index)  
plt.title('Count of Complaints by Root Cause')  
plt.xlabel('Count')  
plt.ylabel('Root Cause')  
plt.show() 
```

```python
plt.figure(figsize=(12, 6))  
sns.countplot(y='Symptom Condition 1', data=df, order=df['Symptom Condition 1'].value_counts().index)  
plt.title('Count of Complaints by Symptom Condition')  
plt.xlabel('Count')  
plt.ylabel('Symptom Condition')  
plt.show()  
```
